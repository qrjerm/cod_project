```python
#Define our path to chromedriver
PATH = "C:\Program Files (x86)\chromedriver.exe" #Version 102.0.5005.63
driver = webdriver.Chrome(PATH)

#Add an implicit wait to avoid interruptions while scraping out data
driver.implicitly_wait(10)

#Define our source of data, in this case, a profile from cod.tracker.gg 
driver.get("https://cod.tracker.gg/modern-warfare/profile/battlenet/ChefRoy%2311358/mp/matches")
#Codtracker loads the last 20 games played and stats for each match so we instruct our driver to scrape those stats
driver.find_elements(by=By.XPATH, value='//*[@id="app"]/div[2]/div[2]/div/main/div[2]/div[3]/div[2]/div/div/div[2]/div[1]/div[2]')


#Our scraped elements are in the following form: "selenium.webdriver.remote.webelement.WebElement" 
#So we will convert the elements into a text file for usability
#We create an empty list (mystats) and append our textfiles using a for loop
main = driver.find_elements(by=By.XPATH, value='//*[@id="app"]/div[2]/div[2]/div/main/div[2]/div[3]/div[2]/div/div/div[2]/div[1]/div[2]')
mystats = []
for value in main:
       print(value.text)
        
for stats in main:
    print(value.text)
    mystats.append(value.text)

#Once our webdriver is finished, the chrome browser will quit
driver.quit()
    
```

```python
#We will eventually be making a table in excel out of our values
#In the meantime we will remove descriptors such as 'View','Kills','Deaths',etc because we will add them in later

OurValues = cleanString
unwanted = {'',' ','View','Kills','Deaths','K/D Ratio','Score/min',
 'Ratio'}
OurValues = [e for e in OurValues if e not in unwanted]

OurValues
#As you can see, the game mode and time are concatenated together
#We could remove the space in between each element
#However, this would also remove the space for elements such as 'Shoot House' which is the full name of a map
#So we will adjust this later in excel using delimiters
```

```python
#Now we will import Pandas and numpy so that we can create a dataframe out of our list
import numpy as np
import pandas as pd
```

```python
input = OurValues

#We convert that to a numpy array as np.array
num = np.array(input)

#Currently its shape is single dimensional, 
#We will change that into a two dimensional matrix with 20 rows and 6 columns
reshaped = num.reshape(20,6)

#We label the headers of our columns and create a dataframe
df = pd.DataFrame(reshaped, columns=['Gamemode & Time','Map','Kills','Deaths','K/D','SPM'])
```

```python
#Take a look at our dataframe
df
```

```python
#Since our elements are still strings, we will convert them into numbers
df["Kills"] = pd.to_numeric(df["Kills"])
df["Deaths"] = pd.to_numeric(df["Deaths"])
df["K/D"] = pd.to_numeric(df["K/D"])
```

```python
#Since we get an error when converting the SPM column, so we ignore errors
pd.to_numeric(df['SPM'], errors='ignore')
```

```python
#Lastly we export our near perfect table to Excel for the final touches!
df.to_excel('CodTrackerData.xlsx',index = False, sheet_name = 'NewSheet' )
```
